```python
#plotting trignometric graphs
```

```python
x_sin=np.arange(0,2*np.pi,0.1)
y_sin=np.sin(x_sin)
print(y_sin)
```

```python
plt.figure(figsize=(3,3))
plt.plot(x_sin,y_sin,'b--',label='sine curve')
plt.title('sin curve')
#provides the title to the graph
plt.legend()
plt.show()
```

```python
x_cos=np.arange(0,2*np.pi,0.1)
y_cos=np.cos(x_cos)

plt.figure(figsize=(3,3))
plt.plot(x_cos,y_cos)
plt.title('cosine curve')
plt.show()
```

```python
x_tan=np.arange(0,2*np.pi,0.1)
y_tan=np.tan(x_tan)

plt.figure(figsize=(3,3))
plt.plot(x_tan,y_tan)
plt.title('tan curve')
plt.show()
```

```python
# subplot
```

```python
plt.figure(figsize=(6,6))

plt.subplot(2,2,1)
plt.plot(x_sin,y_sin,'r-*')
plt.title('sin curve')

plt.subplot(2,2,2)
plt.plot(x_cos,y_cos,'g-o')
plt.title('cos curve')

plt.subplot(2,2,3)
plt.plot(x_tan,y_tan,'g--')
plt.title('tan curve')

plt.show()
```
