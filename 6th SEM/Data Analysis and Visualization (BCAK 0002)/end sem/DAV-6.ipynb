```python
# bokeh library for visualization
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool
```

```python
import numpy as np
# here linspace create a array that starts with 0 and  A
x = np.linspace(0, 4*3.14, 200)
y = np.sin(x)
print (x,'\n',y)
```

```python
#figure creates a window for th graph
p = figure(title='Sine Wave',
           x_axis_label='x',
           y_axis_label='sin(x)',
           width=600, height=350)
```

```python
output_notebook()
```

```python
#add a line
p.line(x, y, line_color='green',
       line_width=2, legend_label='sin(x)')
show(p)
```

```python
# Add hover tooltip
#$x → x-coordinate of mouse position
#$y → y-coordinate of mouse position
#add the hover tool
p.add_tools(HoverTool(tooltips=[('x', '$x'), ('y', '$y')]))
# Add scatter (circle)
show(p)
```

```python
p2 = figure(title='Scatter-Square')
p2.scatter(x, y, size=6, marker="square", color='green')
show(p2)
```

```python
p3 = figure(title='Scatter-circle')
p3.scatter(x, y, size=6, marker="circle", color='blue')
show(p3)
```

```python
import numpy as np

# x axis values (1 to 15)
x = np.arange(1, 16)

#  y axis values (custom data)
y = np.array([12, 18, 15, 22, 28, 24, 30, 35, 33, 40, 38, 45, 50, 48, 55])
```

```python
# Vertical Bar Graph
p4 = figure(title='Vertical Bar')
p4.vbar(x=x, top=y, width=0.3, color='orange')
show(p4)
```

```python
# Horizontal Bar Graph
p5 = figure(title='Horizontal Bar')
p5.hbar(y=y, right=x, height=0.5, color='purple')
show(p5)
```

```python
# Patch (Area)
p6 = figure(title='Patch Graph')
p6.patch(x, y, alpha=0.5, color='pink')
show(p6)
```
