```python
import pandas as pd

# Main DataFrame
df = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'Name': ['Aman', 'Riya', 'Rahul', 'Neha', 'Simran'],
    'City': ['Delhi', 'Mumbai', 'Delhi', 'Mumbai', 'Delhi'],
    'Salary': [50000, 60000, 55000, 65000, 70000],
    'Marks': [85, 90, 78, 88, 95],
    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Jan'],
    'Sales': [200, 300, 250, 400, 350]
})

# Second DataFrame for merge
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Department': ['HR', 'IT', 'Finance', 'Marketing']
})

df2 = pd.DataFrame({
    'ID': [1, 2, 3, 5],
    'Bonus': [5000, 6000, 7000, 8000]
})

print(df)
```

```python
result = df.groupby('City')['Salary'].mean()
print(result)
#groupby() is used to split a DataFrame into groups based columns
#to perfrom operations like sum, mean, count, etc.to each group separately
```

```python
pd.merge(df1, df2, on='ID')
#Combines two datasets (df1 and df2) based on a common column ID.
```

```python
df.pivot_table(values='Sales', index='Month')
#Creates a summary table (like Excel pivot table).
```

```python
df.sort_values('Marks', ascending=False)
#Sorts data based on Marks column in descending order (highest first).
```

```python
df['Grade'] = df['Marks'].apply(
    lambda x: 'A' if x>=85 else 'B')
df
#apply runs the function on every value
#lambda is a way to create a small function without giving it a name
```
