```python
#dropping rows with NaN values
df3=df2.dropna(axis=0) #axis =0 means it removes rows
df3
#removes the rows with null values and stores the data in df3
```

```python
df3.shape
```

```python
df4=df2.dropna(axis=1) #axis =1 means columns  will be removed
df4
```

```python
df4.shape
```

```python
#remove if 'any' row value is null then remove that row
df5=df2.dropna(how='any')
print (df5)
df5.shape
```

```python
df5=df2.dropna(how = 'all')
print (df5)
df5.shape
# remove the row if the enitre row is empty
# in this dataset no row is entirely NULL
```

```python
df2.dropna(inplace = True)
print(df2)
df2.shape
#replace the original dataframe and create a new data frmae that do not have any null values
```
