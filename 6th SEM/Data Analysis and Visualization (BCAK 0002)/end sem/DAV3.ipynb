```python
#filling the Null Vlues
import pandas as pd
```

```python
df=pd.read_csv(r'C:\Users\HP\Downloads\restaurant_data.csv')
df.head()
df
```

```python
df.isnull().sum()
#null value in each column
```

```python
#as null removes the entire row it is costing the data which we can't afford
df.fillna(0)
#fill all the null values with zero
```

```python
df.fillna(2)
#fill all the null values with two
```

```python
df.fillna({'yoy_units':'none', 'segment':0})
#fill the specified column with given values
```

```python
#df.fillna(method='ffill', axis=0) #this might break in future
df.ffill(axis=0) #by default
#the method ffill=forward fill it will fill the null values with the previous row value
```

```python
df.ffill(axis=1)
#the method ffill=forward fill, it will fill the null values with the previous column value
```

```python
df['yoy_units'].fillna(value=df['yoy_units'].mean())
#fill the values of the particular column with mean values
```

```python
#df.fillna(method='bfill')
df.bfill(axis=0)
#fill the null values with the next row value
```

```python
df.bfill(axis=1)
#fill the null values with the next column value
```

```python
#df.fillna(method = 'bfill', inplace = True)
df.bfill(inplace = True)
df
#updates the df directly, no new DataFrame is created
```

```python
#df.replace(to_replace=-0.3, value=-0.4)
#replace a particular va;ue with other value
#shortcut
df.replace(-0.3,-0.4)
```

```python
df =df.drop_duplicates()
#chaeck identical rows and delete them
df
```

```python
#removes rows based on column
df = df.drop_duplicates(subset='segment')
df
```
