```python
import plotly.express as px
# used to import the plotly library
```

```python
# Load built-in dataset
df = px.data.gapminder()
df
```

```python
#want to draw the GDP of india in different years
fig = px.line(df[df['country']=='India'],
              x='year', y='gdpPercap',
              title='India GDP', markers=True)
fig.show()
#here df[df['country']=='India'] filters the dataset and selects only the rows where country is India
```

```python
#gdp wrt life expectancy of the continent
fig = px.scatter(df, x='gdpPercap',
                 y='lifeExp', size='pop',
                 color='continent',
                 animation_frame='year',
                 log_x=True,opacity=0.8, hover_name='country')
fig.show()
#here px.scatter crates a scatter plot
# size is bubble size ='pop' means the larger the population the bigger the bubble
#color (Adds color categories) represents different color for continent
#animation-frame (Creates animation over a column)='year' make the graph move with repect to year
#log_x=True Changes X-axis to logarithmic scale for better scale
#opacity provide opacity of graph
#hover_name Shows name when you hoverd  skjdnpjn thie h rj s,
```

```python
px.bar(df, x='country', y='gdpPercap').show()
px.bar(df, x='country', y='pop').show()
px.bar(df, x='continent', y='gdpPercap').show()
#plots bar graph
```

```python
px.pie(df,values='pop', names='continent').show()
```

```python
px.box(df, x='continent', y='lifeExp', animation_frame='year').show()
```

```python
px.area(df, x='year', y='pop', animation_frame='continent')
```

```python
px.histogram(df,x='lifeExp',animation_frame='year').show()
```

```python
fig = px.scatter_geo(
    df, locations='iso_alpha',
    size='pop', color='continent')
fig.show()
```

```python
import plotly.express as px
# Create line plot for all continents
fig = px.line(df,
              x='year',
              y='lifeExp',
              color='continent',   
              title='Life Expectancy of All Countries Over Years',
              markers=True,)

# Show plot
fig.show()
```

```python
px.choropleth(df, locations='iso_alpha', color='gdpPercap',
              animation_frame='year')
```

```python
df1 = px.data.iris()
df1
```

```python
# Line Chart
fig1 = px.line(df1,
               x='sepal_length',
               y='petal_length',
               title='Sepal Length vs Petal Length',
               markers=True)
fig1.show()

# Box Plot
fig2 = px.box(df1,
              x='species',
              y='sepal_length',
              title='Sepal Length Distribution by Species')
fig2.show()

# Scatter Plot
fig3 = px.scatter(df1,
                  x='sepal_length',
                  y='sepal_width',
                  color='species',
                  title='Sepal Length vs Sepal Width')
fig3.show() 

#bar graph

fig4 = px.bar(df1,
              x='species',
              y='sepal_length',
              title='Sepal Length Distribution by Species').show()
```
